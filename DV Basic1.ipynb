```python
import pandas as pd
import numpy as np
import seaborn as sb
### Import data visualization packages
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Creating the dataset
df=sb.load_dataset("iris")
df.head()
```

```python
df.info()
```

```python
df1=df.groupby('sepal_length')['sepal_width'].sum().to_frame().reset_index()
```

```python
df1.head()
```

```python
plt.plot(df1['sepal_length'],df1['sepal_width'])
#Adding the asthetics
plt.title("sepal length Vs sepal width")
plt.xlabel("sepal length")
plt.ylabel("seapal width")
#show the plot
plt.show()
```

```python
df2=df.groupby('petal_length')['petal_width'].sum().to_frame().reset_index()
df2.head()
```

```python
#Creating the line chart
plt.plot(df2['petal_length'],df2['petal_width'])
#Adding the asthetics
plt.title("petal length Vs petal width")
plt.xlabel("petal length")
plt.ylabel("petal width")
#show the plot
plt.show()
```
