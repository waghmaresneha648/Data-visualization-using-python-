```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris






```

```python
# 1. Load the dataset
iris_data = load_iris()
df = pd.DataFrame(iris_data.data, columns=iris_data.feature_names)
df['species'] = [iris_data.target_names[i] for i in iris_data.target]
```

```python
# 2. Basic Data Inspection
print("--- Dataset Info ---")
print(df.info())
print("\n--- Descriptive Statistics ---")
print(df.describe())
print("\n--- Class Distribution ---")
print(df['species'].value_counts())
```

```python
# 3. Visualization
sns.set_theme(style="whitegrid")

# Pairplot to see relationships between all features
plt.figure(figsize=(10, 8))
sns.pairplot(df, hue='species', palette='viridis', markers=["o", "s", "D"])
plt.suptitle("Pairwise Relationships of Iris Features", y=1.02)
plt.show()
```
