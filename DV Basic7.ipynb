```python


import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
cols = ['CreditScore','Geography','Gender','Age','Tenure',
        'Balance','NumOfProducts','IsActiveMember','EstimatedSalary','Exited']

churn = pd.read_csv(
    r"C:\Users\LENOVO\Downloads\archive (2)\Churn_Modelling.csv",
    usecols=cols
)

churn['Status'] = churn['Exited'].map({0: 'Stayed', 1: 'Exited'})
churn['IsActiveMember'] = churn['IsActiveMember'].map({0: 'Inactive', 1: 'Active'})
churn['NumOfProducts'] = churn['NumOfProducts'].astype(str) + ' Product(s)'

churn.head()
```

```python
geo = churn['Geography'].value_counts().reset_index()
geo.columns = ['Country', 'Count']
fig = px.bar(geo, x='Country', y='Count',
             color='Country',
             color_discrete_sequence=px.colors.qualitative.Set2,
             text='Count',
             title='Customers by Geography')

fig.update_traces(textposition='outside')
fig.update_layout(showlegend=False, yaxis_title='Number of Customers',
                  title_font_size=16)
fig.show()
```

```python
grp = (churn.groupby(['Geography', 'Status'])
        .size().reset_index(name='Count'))

fig = px.bar(grp, x='Geography', y='Count', color='Status',
             barmode='group',
             color_discrete_map={'Stayed': '#4c8cbf', 'Exited': '#e05c5c'},
             text='Count',
             title='Stayed vs Exited by Geography')

fig.update_traces(textposition='outside')
fig.update_layout(yaxis_title='Number of Customers', title_font_size=16,
                  legend_title='Status')
fig.show()
```

```python
status_counts = churn['Status'].value_counts().reset_index()
status_counts.columns = ['Status', 'Count']

fig = px.pie(status_counts, names='Status', values='Count',
             hole=0.45,
             color='Status',
             color_discrete_map={'Stayed': '#4c8cbf', 'Exited': '#e05c5c'},
             title='Overall Churn Rate')

fig.update_traces(textinfo='percent+label', pull=[0, 0.05])
fig.update_layout(title_font_size=16, legend_title='Status')
fig.show()
```

```python
sample = churn.sample(800, random_state=42)

fig = px.scatter(sample, x='Age', y='Balance',
                 color='Status',
                 color_discrete_map={'Stayed': '#4c8cbf', 'Exited': '#e05c5c'},
                 size='EstimatedSalary',
                 size_max=14,
                 hover_data=['CreditScore', 'Geography', 'Gender', 'Tenure'],
                 opacity=0.65,
                 title='Age vs Balance (sized by Salary, coloured by Churn)')

fig.update_layout(title_font_size=15,
                  yaxis_tickprefix='$', yaxis_tickformat=',',
                  legend_title='Status')
fig.show()
```

```python
fig = px.box(churn, x='Geography', y='CreditScore',
             color='Status',
             color_discrete_map={'Stayed': '#4c8cbf', 'Exited': '#e05c5c'},
             points='outliers',
             notched=True,           # notch shows 95% CI around the median
             title='Credit Score by Geography and Churn Status')

fig.update_layout(title_font_size=15, legend_title='Status',
                  yaxis_title='Credit Score')
fig.show()
```

```python
num_cols = ['CreditScore', 'Age', 'Tenure', 'Balance',
             'NumOfProducts', 'EstimatedSalary', 'Exited']

corr = churn[num_cols].corr().round(2)

fig = px.imshow(corr,
                text_auto=True,
                color_continuous_scale='RdYlGn',
                zmin=-1, zmax=1,
                aspect='auto',
                title='Correlation Matrix — Churn Dataset')

fig.update_layout(title_font_size=15,
                  coloraxis_colorbar_title='r')
fig.show()
```

```python
avg = (churn.groupby(['Age', 'Status'])['Balance']
       .mean().reset_index())

fig = px.line(avg, x='Age', y='Balance', color='Status',
              color_discrete_map={'Stayed': '#4c8cbf', 'Exited': '#e05c5c'},
              markers=True,
              title='Average Balance by Age and Churn Status')

fig.update_layout(
    title_font_size=15,
    yaxis_tickprefix='$', yaxis_tickformat=',',
    legend_title='Status',
    xaxis=dict(rangeslider=dict(visible=True), type='linear')
)
fig.show()
```

```python
sun = (churn.groupby(['Geography', 'Status', 'NumOfProducts'])
       .size().reset_index(name='Count'))

fig = px.sunburst(sun,
                  path=['Geography', 'Status', 'NumOfProducts'],
                  values='Count',
                  color='Status',
                  color_discrete_map={'Stayed': '#4c8cbf',
                                      'Exited': '#e05c5c',
                                      '(?)':    '#cccccc'},
                  title='Customers: Geography → Churn Status → # Products')

fig.update_layout(title_font_size=15)
fig.show()
```

```python
total         = len(churn)
active        = churn['IsActiveMember'].eq('Active').sum()
multi_product = churn['NumOfProducts'].isin(['2 Product(s)', '3 Product(s)', '4 Product(s)']).sum()
high_credit   = churn['CreditScore'].ge(700).sum()
stayed        = churn['Status'].eq('Stayed').sum()

funnel_df = pd.DataFrame({
    'Stage': ['All Customers', 'Active Members',
               'Multi-Product', 'Credit Score ≥ 700', 'Stayed (Did Not Churn)'],
    'Count': [total, active, multi_product, high_credit, stayed]
})

fig = px.funnel(funnel_df, x='Count', y='Stage',
                color_discrete_sequence=['#4c8cbf'],
                title='Customer Retention Funnel')

fig.update_layout(title_font_size=15)
fig.show()
```
