```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cols=['CreditScore','Geography',
'Gender','Age','Tenure',
'Balance','NumOfProducts',
'IsActiveMember','EstimatedSalary','Exited']
churn=pd.read_csv("C:\\Users\\LENOVO\\Downloads\\archive (2)\\Churn_Modelling.csv",usecols=cols)
churn.head()
churn.tail()
```

```python
plt.figure(figsize=(8,5))
plt.title("Number of Customers",fontsize=14)
plt.bar(x=churn['Geography'].value_counts().index,
height=churn.Geography.value_counts().values)
```

```python
plt.figure(figsize=(8,5))
plt.title("Number of Customers",fontsize=14)
plt.bar(x=churn['Geography'].value_counts().index,
height=churn.Geography.value_counts().values)
plt.xticks(fontsize=12,rotation=45)
plt.yticks(ticks=np.arange(0,7000,1000),fontsize=12)
```
